```python
import control as ctrl 
import numpy as np
import matplotlib.pyplot as plt

from utils import *

s = ctrl.tf('s')
T1 = pt2(100,0.2)
T2 = pt2(150,0.05)
ctrl.bode(T1)
ctrl.bode(T2)
```

```python


S1 = 1 - T1
S2 = 1 - T2

S1.name = 'Se1'
S2.name = 'Se2'

We = s/(s + 70) * 4
We.name = 'We'

wlim = [10,1000]
ctrl.bode(S1,omega_limits=wlim)
ctrl.bode(S2,omega_limits=wlim)
ctrl.bode(We,omega_limits=wlim,linestyle='--',color='k')

save_fig('hinf_scale_ex0')

plt.figure()

S1w = S1/We
S2w = S2/We

S1w.name = 'S1/We'
S2w.name = 'S2/We'

ctrl.bode(S1w,omega_limits=wlim)
ctrl.bode(S2w,omega_limits=wlim)
save_fig('hinf_scale_ex1')
```
